<h2>Problem Statement</h2>


<h3>Context</h3><p>AllLife Bank is a US bank that has a growing customer base. The majority of these customers are liability customers (depositors) with varying sizes of deposits. The number of customers who are also borrowers (asset customers) is quite small, and the bank is interested in expanding this base rapidly to bring in more loan business and in the process, earn more through the interest on loans. In particular, the management wants to explore ways of converting its liability customers to personal loan customers (while retaining them as depositors).</p>
<p>A campaign that the bank ran last year for liability customers showed a healthy conversion rate of over 9% success. This has encouraged the retail marketing department to devise campaigns with better target marketing to increase the success ratio.</p>
<p>You as a Data scientist at AllLife bank have to build a model that will help the marketing department to identify the potential customers who have a higher probability of purchasing the loan.</p>


<h3>Objective</h3><p>To predict whether a liability customer will buy personal loans, to understand which customer attributes are most significant in driving purchases, and identify which segment of customers to target more.</p>


<h3>Data Dictionary</h3><ul>
<li><code>ID</code>: Customer ID</li>
<li><code>Age</code>: Customer’s age in completed years</li>
<li><code>Experience</code>: #years of professional experience</li>
<li><code>Income</code>: Annual income of the customer (in thousand dollars)</li>

```python
# Calculate the median of non-negative experience values
median_experience = data[data['Experience'] >= 0]['Experience'].median()

# Replace negative experience values with the median
data['Experience'] = data['Experience'].apply(lambda x: median_experience if x < 0 else x)

# Verify that there are no more negative values
print("Min Experience after handling negative values:", data['Experience'].min())
```

```python
# Identify numerical features for outlier detection
numerical_features_for_outliers = ['Income', 'CCAvg', 'Mortgage']

plt.figure(figsize=(15, 5))

for i, col in enumerate(numerical_features_for_outliers):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(data=data, y=col)
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers for the specified numerical features
outliers_income = detect_outliers_iqr(data, 'Income')
outliers_ccavg = detect_outliers_iqr(data, 'CCAvg')
outliers_mortgage = detect_outliers_iqr(data, 'Mortgage')

print("Outliers in Income:")
display(outliers_income)

print("\nOutliers in CCAvg:")
display(outliers_ccavg)

print("\nOutliers in Mortgage:")
display(outliers_mortgage)
```

<p>Even though there are some outliers , I am not going to treat them because Decision Trees are less susceptible to outliers.</p>

<p>I have defined a couple of methods below that I will be resuing for different models.</p>


```python
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, f1_score
import pandas as pd # Import pandas here if not already imported

def get_classification_report(y_true, y_pred, name):

    acc = accuracy_score(y_true, y_pred)  # to compute Accuracy
    recall = recall_score(y_true, y_pred)  # to compute Recall
    precision = precision_score(y_true, y_pred)  # to compute Precision
    f1 = f1_score(y_true, y_pred)  # to compute F1-score

    # creating a dataframe of metrics with metrics as index
    df_perf = pd.DataFrame(
        {'Metrics': ['Accuracy', 'Recall', 'Precision', 'F1'],
         name: [acc, recall, precision, f1]}
    ).set_index('Metrics')

    return df_perf
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Confusion Matrix - {title}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

def plot_decision_tree(model, feature_names, class_names=['Not Accepted', 'Accepted'], figsize=(20, 10), fontsize=10):
    plt.figure(figsize=figsize)
    plot_tree(model, feature_names=feature_names, filled=True, rounded=True, class_names=class_names, fontsize=fontsize)
    plt.title("Decision Tree Visualization")
    plt.show()
```

<h3>Model Building</h3>


```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion="gini", random_state=42)
model.fit(X_train, y_train)
```

```python
# Generate predictions on the training set
y_train_pred = model.predict(X_train)

# Get metrics for the Decision Tree model on the training set
dt_train_metrics = get_classification_report(y_train, y_train_pred, "Decision Tree (Train)")
print("Decision Tree (Train) Metrics:")
display(dt_train_metrics)

plot_confusion_matrix(y_train, y_train_pred, "Train Set")
```

<p>All the metrics seem to be 1 on the training set and suggests overfiting possibly.</p>


```python
# Generate predictions on the test set
y_test_pred = model.predict(X_test)

# Get metrics for the Decision Tree model on the test set
dt_test_metrics = get_classification_report(y_test, y_test_pred, "Decision Tree (Test)")
print("\nDecision Tree (Test) Metrics:")
display(dt_test_metrics)

# Plot confusion matrix for the test set
plot_confusion_matrix(y_test, y_test_pred, "Test Set")
```

<p>The test metrics are decently high with recall being 89% and precision being 94%.</p>


```python
plot_decision_tree(model, X_train.columns.tolist(), figsize=(30, 15), fontsize=8)
```

<h2>Model Performance Improvement</h2>


<h3>Pre Pruning</h3>


```python
# Define the parameters of the tree to iterate over
max_depth_values = np.arange(2, 10, 2)
max_leaf_nodes_values = [50, 75, 150, 250]
min_samples_split_values = [10, 30, 50, 70]

# Initialize variables to store the best model and its performance
best_estimator = None
best_score_diff = float('inf')
best_test_score = 0.0

# Iterate over all combinations of the specified parameter values
for max_depth in max_depth_values:
    for max_leaf_nodes in max_leaf_nodes_values:
        for min_samples_split in min_samples_split_values:

            # Initialize the tree with the current set of parameters
            estimator = DecisionTreeClassifier(
                max_depth=max_depth,
                max_leaf_nodes=max_leaf_nodes,
                min_samples_split=min_samples_split,
                class_weight='balanced', # Corrected parameter name
                random_state=42
            )

            # Fit the model to the training data
            estimator.fit(X_train, y_train)

            # Make predictions on the training and test sets
            y_train_pred = estimator.predict(X_train)
            y_test_pred = estimator.predict(X_test)

            # Calculate recall scores for training and test sets
            train_recall_score = recall_score(y_train, y_train_pred)
            test_recall_score = recall_score(y_test, y_test_pred)

            # Calculate the absolute difference between training and test recall scores
            score_diff = abs(train_recall_score - test_recall_score)

            # Update the best estimator and best score if the current one has a smaller score difference
            if (score_diff < best_score_diff) & (test_recall_score > best_test_score):
                best_score_diff = score_diff
                best_test_score = test_recall_score
                best_estimator = estimator

# Print the best parameters
print("Best parameters found:")
print(f"Max depth: {best_estimator.max_depth}")
print(f"Max leaf nodes: {best_estimator.max_leaf_nodes}")
print(f"Min samples split: {best_estimator.min_samples_split}")
print(f"Best test recall score: {best_test_score}")
```

<p>It looks like after pre pruning my tree with different params, the best max depth is 2, max leaf nodes are 50 and min samples split is 10. Here we have used recall score to find the best hyper parameters.</p>


```python
# Generate predictions on the training set using the best estimator
y_train_pred_best = best_estimator.predict(X_train)

# Get metrics for the best Decision Tree model on the training set
best_dt_train_metrics = get_classification_report(y_train, y_train_pred_best, "Pre Pruned Tree  (Train)")
print("Pre Pruned Tree (Train) Metrics:")
display(best_dt_train_metrics)

# Plot confusion matrix for the training set using the best model
plot_confusion_matrix(y_train, y_train_pred_best, "Best Model Train Set")
```

<p>After pre pruning, it looks like our metrics seem to have dropped significantly but recall value is 1 indicating a good value especially for our use case.</p>


```python
# Evaluate the best model found during hyperparameter tuning on the test set

# Generate predictions on the test set using the best estimator
y_test_pred_best = best_estimator.predict(X_test)

# Get metrics for the best Decision Tree model on the test set
best_dt_test_metrics = get_classification_report(y_test, y_test_pred_best, "Pre Pruned Tree (Test)")
print("Pre Pruned Tree  (Test) Metrics:")
display(best_dt_test_metrics)

# Plot confusion matrix for the test set using the best model
plot_confusion_matrix(y_test, y_test_pred_best, "Best Model Test Set")
```

<p>On the training data set also after pre pruning we have recall of 1 but accuracy is 79%.</p>


```python
# Print the optimal hyperparameters found for the best model
print("Optimal Hyperparameters for the Best Model:")
print(f"  Max Depth: {best_estimator.max_depth}")
print(f"  Min Samples Split: {best_estimator.min_samples_split}")
print(f"  Min Samples Leaf: {best_estimator.min_samples_leaf}")

# Get feature importances from the best_model
feature_importances = pd.Series(best_estimator.feature_importances_, index=X_train.columns)

# Sort feature importances in descending order
sorted_feature_importances = feature_importances.sort_values(ascending=False)

# Print the sorted feature importances
print("\nSorted Feature Importances:")
display(sorted_feature_importances)
```

<h3>Post Pruning</h3>


```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import recall_score

# 1. Calculate the cost complexity pruning path
clf = DecisionTreeClassifier(random_state=42)
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Optional: Visualize Impurity vs. alpha
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(ccp_alphas[:-1], impurities[:-1], marker="o", drawstyle="steps-post")
ax.set_xlabel("effective alpha")
ax.set_ylabel("total impurity of leaves")
ax.set_title("Total Impurity vs effective alpha for training set")
plt.show()

# 2. Train a Decision Tree for each effective alpha
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)     ## Complete the code to fit decision tree on training data
    clfs.append(clf)

# Remove the last tree (single-node tree) and its alpha
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]

# 3. Analyze Tree Properties vs. alpha
node_counts = [clf.tree_.node_count for clf in clfs]
depth = [clf.tree_.max_depth for clf in clfs]

fig, ax = plt.subplots(2, 1, figsize=(10, 7))
ax[0].plot(ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")
ax[1].plot(ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")
fig.tight_layout()
plt.show()

# 4. Evaluate Recall vs. alpha for training and testing sets
recall_train = []
recall_test = []
for clf in clfs:
    pred_train = clf.predict(X_train)
    values_train = recall_score(y_train, pred_train)
    recall_train.append(values_train)

    pred_test = clf.predict(X_test)
    values_test = recall_score(y_test, pred_test)
    recall_test.append(values_test)

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_xlabel("alpha")
ax.set_ylabel("Recall")
ax.set_title("Recall vs alpha for training and testing sets")
ax.plot(ccp_alphas, recall_train, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, recall_test, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
# Get the ccp_alpha value corresponding to the maximum test recall
optimal_ccp_alpha_by_recall = 0.005

print(f"The optimal ccp_alpha based on maximizing test recall is: {optimal_ccp_alpha_by_recall:.6f}")
```

<p>Based on the above image I have chosen the optimal ccp_alpha as 0.005. I also reviwed with 0.0074.</p>


```python
# Train the final Decision Tree model with the optimal ccp_alpha found by maximizing test recall
final_post_pruned_model = DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_ccp_alpha_by_recall)

# Train the model on the training data
final_post_pruned_model.fit(X_train, y_train)

# Evaluate the final post-pruned model on the test set
y_test_pred_final_post_pruned = final_post_pruned_model.predict(X_test)

# Get metrics for the final post-pruned model on the test set
final_post_pruned_test_metrics = get_classification_report(y_test, y_test_pred_final_post_pruned, "Post-Pruned Tree (Test)")
print("Post-Pruned Decision Tree (Test) Metrics:")
display(final_post_pruned_test_metrics)

# Plot confusion matrix for the test set using the final post-pruned model
plot_confusion_matrix(y_test, y_test_pred_final_post_pruned, "Post-Pruned Tree Test Set")
```
